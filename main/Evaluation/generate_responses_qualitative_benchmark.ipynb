```python
import time

def answer(question,model_,tokens):
    start_time = time.time()
    # Define the new task for the assistant
    prompt = (
        f"{question}"
    )

    # Update system message to match the new task
    system_message = "You are an assistant that that responds to a question with an answer."

    # Use OpenAI API to generate the response
    response = client.chat.completions.create(
        model=model_,  # Adjust the model as per availability or requirements
        messages=[
            {
                "role": "system",
                "content": system_message
            },
            {
                "role": "user",
                "content": prompt
            }
        ],
        temperature=0.7,  # Adjust temperature as needed to balance creativity and accuracy
        max_tokens=tokens,  # Adjust max tokens if more are needed for longer answers
        top_p=1
    )

    complete_response = response.choices[0].message.content.strip()
    tokens_input = response.usage.prompt_tokens
    tokens_output = response.usage.completion_tokens
    end_time = time.time()  # Capture end time after the response has been generated
    # Calculate the duration in seconds
    duration = end_time - start_time
    duration= round(duration, 2)
    # Return the generated response
    return [complete_response, tokens_input, tokens_output, duration]
```

```python

import torch
from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig

base_model_id = "mistralai/Mistral-7B-v0.1"
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.bfloat16
)

model = AutoModelForCausalLM.from_pretrained(base_model_id, quantization_config=bnb_config, device_map="auto")
```

```python
# Init an eval tokenizer that doesn't add padding or eos token
eval_tokenizer = AutoTokenizer.from_pretrained(
    base_model_id,
    add_bos_token=True,
)
```

```python
import pandas as pd
df_questions=pd.read_csv('/content/drive/MyDrive/Arunadiss/main/Evaluation/Data/Clean Data/qualitative_data_questions/qualitative_questions.csv')
df_questions=df_questions[['Qualitative benchmark question','Markscheme']]
df_questions['Qualitative benchmark question'][0]
```

```python
instruction_prompt='You are an assistant that that responds to a question with an answer. You will be provided with a question. You must stop after providing your answer.'
```

```python
import time

def answer_vanilla_mistral(question,tokens):
    start_time = time.time()
    #question=instruction_prompt+'\n\nQuestion: '+question+'\n\n Answer:'
    question=instruction_prompt+'\n\n'+question+'\n\nAnswer:'

    model_input = eval_tokenizer(question, return_tensors="pt").to("cuda")
    n_input_tokens=len(model_input['input_ids'][0])
    model.eval()
    with torch.no_grad():
        answer=eval_tokenizer.decode(model.generate(**model_input, max_new_tokens=tokens, repetition_penalty=1.15, pad_token_id=eval_tokenizer.eos_token_id)[0], skip_special_tokens=True)
    idx=len(question)+1
    answer=answer[idx:]
    tokenize_answer = eval_tokenizer(answer, return_tensors="pt").to("cuda")
    n_output_tokens=len(tokenize_answer['input_ids'][0])
    end_time = time.time()  # Capture end time after the response has been generated
    # Calculate the duration in seconds
    duration = end_time - start_time
    duration= round(duration, 2)
    #give me everything after question from my answer string
    return [answer,n_input_tokens,n_output_tokens,duration]
```

```python
from peft import PeftModel

ft_model = PeftModel.from_pretrained(base_model, "/content/drive/MyDrive/Arunadiss/main/Finetune/model/mistral/logs/mistral-v3-ib-bio-v1/checkpoint-1500")
```

```python
import pandas as pd
df_questions=pd.read_csv('/content/drive/MyDrive/Arunadiss/main/Evaluation/Data/Clean Data/qualitative_data_questions/qualitative_questions.csv')
df_questions=df_questions[['Qualitative benchmark question','Markscheme']]
df_questions['Qualitative benchmark question'][0]
```

```python
instruction_prompt="""
You are an AI assistant designed to help learners using the INSPIRE framework, which emphasizes Intelligent, Nurturant, Socratic, Progressive, Indirect, Reflective, and Encouraging approaches in tutoring.

- As an Intelligent tutor, use your vast knowledge base to provide detailed explanations, incorporate relevant historical context, employ visual aids, and draw real-world analogies to clarify concepts.

- Be Nurturant by showing empathy towards learners' struggles, offering supportive feedback, and encouraging perseverance and resilience.

- Adopt a Socratic approach by asking open-ended questions that encourage critical thinking and exploration, rather than providing direct answers.

- Ensure a Progressive learning experience by adjusting the complexity of questions and topics based on the learner's progress and responses, gradually challenging them with more difficult material.

- Provide Indirect guidance by offering hints and nudges towards solutions instead of outright answers, encouraging learners to think independently and develop problem-solving skills.

- Foster Reflective thinking by prompting learners to verbalize their thought processes, reflect on their answers, and learn from both their successes and mistakes.

- Be Encouraging by highlighting learners' achievements, motivating them to tackle challenging problems, and maintaining a positive, empowering learning environment.

You will be provided with a question.
"""
```

```python
import time


def answer_ft_mistral(question,tokens):
    start_time = time.time()
    question=instruction_prompt+'\n\nQuestion: '+question+'\n\n Answer:'
    #question=instruction_prompt+'\n\n'+question+'\n\nAnswer:'

    model_input = eval_tokenizer(question, return_tensors="pt").to("cuda")
    n_input_tokens=len(model_input['input_ids'][0])
    ft_model.eval()
    with torch.no_grad():
        answer=eval_tokenizer.decode(ft_model.generate(**model_input, max_new_tokens=tokens, repetition_penalty=1.15, pad_token_id=eval_tokenizer.eos_token_id)[0], skip_special_tokens=True)
    idx=len(question)+1
    answer=answer[idx:]
    tokenize_answer = eval_tokenizer(answer, return_tensors="pt").to("cuda")
    n_output_tokens=len(tokenize_answer['input_ids'][0])
    end_time = time.time()  # Capture end time after the response has been generated
    # Calculate the duration in seconds
    duration = end_time - start_time
    duration= round(duration, 2)
    #give me everything after question from my answer string
    return [answer,n_input_tokens,n_output_tokens,duration]
```
