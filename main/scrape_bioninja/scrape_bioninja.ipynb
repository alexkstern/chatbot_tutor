```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import re
from reportlab.lib.pagesizes import letter
from reportlab.lib.styles import getSampleStyleSheet
from reportlab.platypus import Paragraph, SimpleDocTemplate, Spacer
import os
```

```python
URL='https://old-ib.bioninja.com.au/'
page = requests.get(URL)
soup = BeautifulSoup(page.content, "html.parser")
links=[]
titles_links=[]
for a_href in soup.find_all("a", href=True, title=True):
    titles_links.append(a_href["title"])
    links.append(a_href['href'])

# Convert the relative links to absolute URLs
absolute_urls = [URL + link for link in links]
```

```python
df_web_scrape = pd.DataFrame({'title': titles_links, 'urls': absolute_urls})

pages_to_scrape= [
 '1.1 Introduction to Cells',
 '1.2 Ultrastructure of Cells',
 '1.3 Membrane Structure',
 '1.4 Membrane Transport',
 '1.5 The Origin of Cells',
 '1.6 Cell Division',
 '2.1 Molecules to Metabolism',
 '2.2 Water',
 '2.3 Carbohydrates and Lipids',
 '2.4 Proteins',
 '2.5 Enzymes',
 '2.6 Structure of DNA and RNA',
 '2.7 DNA Replication, Transcription & Translation',
 '2.8 Cell Respiration',
 '2.9 Photosynthesis',
 '3.1 Genes',
 '3.2 Chromosomes',
 '3.3 Meiosis',
 '3.4 Inheritance',
 '3.5 Genetic Modification and Biotechnology',
 '4.1 Species, Communities and Ecosystems',
 '4.2 Energy Flow',
 '4.3 Carbon Cycling',
 '4.4 Climate Change',
 '5.1 Evidence for Evolution',
 '5.2 Natural Selection',
 '5.3 Classification of Biodiversity',
 '5.4 Cladistics',
 '6.1 Digestion and Absorption',
 '6.2 The Blood System',
 '6.3 Defence Against Infectious Disease',
 '6.4 Gas Exchange',
 '6.5 Neurons and Synapses',
 '6.6 Hormones, Homeostasis and Reproduction',
 '7.1 DNA Structure and Replication',
 '7.2 Transcription and Gene Expression',
 '7.3 Translation',
 '8.1 Metabolism',
 '8.2 Cell Respiration',
 '8.3 Photosynthesis',
 '9.1 Transport in the Xylem of Plants',
 '9.2 Transport in the Phloem of Plants',
 '9.3 Growth in Plants',
 '9.4 Reproduction in Plants',
 '10.1 Meiosis',
 '10.2 Inheritance',
 '10.3 Gene Pools and Speciation',
 '11.1 Antibody Production and Vaccination',
 '11.2 Movement',
 '11.3 The Kidney and Osmoregulation',
 '11.4 Sexual Reproduction']


for i in range(len(df_web_scrape.title)):
    if df_web_scrape.title[i] not in pages_to_scrape:
        df_web_scrape=df_web_scrape.drop(i)


df_web_scrape=df_web_scrape.reset_index()
df_web_scrape=df_web_scrape.drop('index', axis=1)
df_web_scrape
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from tqdm import tqdm

# Initialize lists to store the results
links_subtopic = []
title_links_subtopic = []
subtopic_of_title = []

# Use tqdm in the loop to display a progress bar
for base_url, title in tqdm(zip(df_web_scrape['urls'], df_web_scrape['title']), total=df_web_scrape.shape[0]):
    # Fetch the HTML code of the webpage
    response = requests.get(base_url)
    soup = BeautifulSoup(response.content, 'html.parser', from_encoding=response.encoding)

    # Use list comprehension to filter 'img' tags with an 'a' parent
    images_with_links = [(img, img.findParent('a')) for img in soup.findAll('img') if img.findParent('a')]

    for img, parent in images_with_links:
        # Extract the hyperlink and resolve the relative URL to a full URL
        href = parent['href']
        href_url = requests.compat.urljoin(base_url, href)

        # Extracting the name of the subtopic from the href attribute
        result = href.split('.html')[0].split('/')[-1].replace('-', ' ')

        # Appending everything to the lists
        links_subtopic.append(href_url)
        title_links_subtopic.append(title)
        subtopic_of_title.append(result)


df_subtopics = pd.DataFrame({
    'Base Title': title_links_subtopic,
    'Subtopic Title': subtopic_of_title,
    'URL': links_subtopic
})

df_subtopics
```

```python
#Remove rows that we do not need
df_subtopics = df_subtopics[df_subtopics['Subtopic Title'] != '']
```

```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
from tqdm import tqdm


# Initialize lists to store the results
titles = []
subtopics = []
prompts = []
completions = []

# Iterate over each row in the DataFrame with tqdm for progress tracking
for index, row in tqdm(df_subtopics.iterrows(), total=df_subtopics.shape[0]):
    URL = row['URL']
    response = requests.get(URL)
    soup = BeautifulSoup(response.content, "html.parser")

    # Process callouts
    callouts = soup.find_all("div", class_="callout")
    callout_texts = [callout.text.strip().replace('\n', '') for callout in callouts if callout.text.strip() != '']
    prompts.extend(callout_texts if callout_texts else ['NaN'])

    # Process article content for specific criteria
    articlesContent = soup.find_all("div", class_="article-content")
    text = ''
    for article in articlesContent:
        spans = article.find_all("span")
        for span in spans:
            if any(keyword in span.text for keyword in ["•", "Understanding:", "Application:", "Skill:"]):
                text += span.text.strip()
            else:
                text += f'{span.text.strip()} '

    # Split text based on specific keywords and filter out empty items
    split_text = re.split(r'Skill:•|Understanding:•|Application:•', text)
    filtered_text = [item.strip() for item in split_text if item.strip() != '']
    completions.extend(filtered_text)

    # Append title and subtopic for each completion
    titles.extend([row['Base Title']] * len(filtered_text))
    subtopics.extend([row['Subtopic Title']] * len(filtered_text))

# Create a DataFrame from the extracted data
df_extracted_features = pd.DataFrame({
    'Title': titles,
    'Subtopic': subtopics,
    'Prompt': prompts,
    'Completion': completions
})

display(df_extracted_features)
```

#Format the scraped dataset

```python
import pandas as pd

df=pd.read_csv('/content/drive/MyDrive/Arunadiss/main/scrape_bioninja/bioninja.csv')
```

```python
# Function to format the strings
def format_chunk(row):
    chunk = f"{row['Title']}\n{row['Subtopic'].capitalize()}:\n"
    if pd.notnull(row['Prompt']):  # Check if Prompt is not NaN
        chunk += f"{row['Prompt']}\n"
    chunk += f"{row['Completion']}'"
    return chunk

# Applying the function to create the new column
df['Chunk'] = df.apply(format_chunk, axis=1)

df
```

```python
# Function to extract the first three characters from the Subtopic column
def extract_subtopic_number(row):
    return row['Title'][:4]

# Applying the function to create the new column 'Subtopic Number'
df['Subtopic Number'] = df.apply(extract_subtopic_number, axis=1).astype(float)

#Extract topic number
def extract_topic_number(row):
    numeric_part = row['Subtopic Number']
    return int(numeric_part)

# Applying the function to create the new column 'Topic Number'
df['Topic Number'] = df.apply(extract_topic_number, axis=1).astype(int)

```

```python
df=df[['Topic Number','Subtopic Number','Chunk']]
df
```
